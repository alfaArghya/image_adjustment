```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = io.imread("1155168.jpg") #importing the img
```

```python
#print(img) #image matrix
```

```python
print(img.shape)
```

```python
#displying the image
plt.imshow(img)
plt.show()
```

```python
cropped = img[400:1350, 1000:1800]  #image croping using numpy slicing
plt.imshow(cropped)
plt.show()
```

```python
io.imsave("img_cropped.jpg",cropped) #save the image
```

```python
mirror_img = img[:,::-1] #creating a mirror image
plt.imshow(mirror_img)
plt.show()
```

```python
io.imsave("mirror_img.jpg",mirror_img) #saving the mirror image
```

```python
red = np.zeros(img.shape, dtype = 'uint8') #creating a numpy 3D array with zeros
red.shape
```

```python
#print(red)
```

```python
#display the image with only red color
red[:,:,0] = img[:,:,0] #store the red channel of image into 3D zero array
plt.imshow(red)
plt.show()
```

```python
#display the image with only green color
green = np.zeros(img.shape, dtype = "uint8") #creating a numpy 3D array with zeros
green[:,:,1] = img[:,:,1] #store the green channel of image into 3D zero array
plt.imshow(green)
plt.show()
```

```python
#display the image with only blue color
blue = np.zeros(img.shape, dtype = "uint8") #creating a numpy 3D array with zeros
blue[:,:,2] = img[:,:,2] #store the blue channel of image into 3D zero array
plt.imshow(blue)
plt.show()
```

```python
RGB = np.vstack((red,green,blue))
plt.imshow(RGB)
plt.show()
```
